Rosenblat's Neuron : Instead of using fixed weight with no learning scope, he introduced random weights and bias with learning scope

```
#assign a random threshold value
act_function= lambda x: 0 if x<=1.5 else 1
train_data=[(array([0,0,1]),0),(array([0,1,1]),1),(array([1,0,1]),1),(array([1,1,1]),0)]
#random.seed(10)
w=random.rand(3)
#w=np.array([0.51416544, 0.34653827, 0.01031072])
eta=0.1
```

```
print(w)
```

```
train_data
```

```
x,expected=choice(train_data)
print(x,expected)
```

```
dot(w,x)
```

```
act_function(dot(w,x))
```

```
#Stochastic Gradient Descent
errors=[]
n=50000
#Training
for i in range(n):
  x,expected=choice(train_data) #replace the choice function and access the data in order (1 to 4)
  net=dot(w,x)   #sum of products
  y_pred=act_function(net)  #predicted output
  error=expected-y_pred   # individual error gradient
  errors.append(error)
  w+=eta*error*x
```

```
w
```

```
#Full Batch Gradient Descent
epoch=50000
#Training
for i in range(epoch):
  error_grad=0
  for j in range(len(train_data)):
    x,expected=train_data[j]
    net=dot(w,x)   #sum of products
    y_pred=act_function(net)  #predicted output
    error=expected-y_pred   # individual error gradient
    error_grad+=error
  w+=eta*error_grad*x
```

```
#Testing
for x,_ in train_data:
  net=dot(x,w)
  print("{} : {} -> {}".format(x[:3],net,act_function(net)))
```
