```python
from sklearn.neural_network import MLPRegressor,MLPClassifier
import scikeras
#for classification problem use KerasClassifier wrapper
from scikeras.wrappers import KerasRegressor
from sklearn.pipeline import Pipeline
```

```python
from sklearn.model_selection import cross_val_score,KFold
```

```python
#Customized GridSearch (not recommended)
scores_mean=[]
scores_sd=[]
for i in np.arange(1,60):
  for j in np.arange(1,60):
    model=MLPRegressor(hidden_layer_sizes=(i,j))
    KF=KFold(n_splits=3,shuffle=True,random_state=0)
    scores=cross_val_score(model,X_std,Y,cv=KF,scoring='neg_root_mean_squared_error')
    scores_mean.append(np.mean(np.abs(scores)))
    scores_sd.append(np.std(np.abs(scores),ddof=1))
print(scores_mean)
print(scores_sd)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(20, input_dim=13,activation='sigmoid'))
model.add(Dense(20,activation='sigmoid'))
model.add(Dense(40,activation='sigmoid'))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer='sgd')
```

```python
model.fit(X_std,Y,batch_size=20,epochs=200,verbose=1)
```

```python
def DL_model2():
  model=Sequential()
  model.add(Dense(50, input_dim=13,activation='sigmoid'))
  model.add(Dense(50,activation='sigmoid'))
  model.add(Dense(40,activation='sigmoid'))
  model.add(Dense(1))
    # Compile model
  model.compile(loss='mean_squared_error', optimizer='sgd')
  return model
```

```python
def DL_model3():
  model=Sequential()
  model.add(Dense(50, input_dim=13,activation='relu'))
  model.add(Dense(50,activation='relu'))
  model.add(Dense(40,activation='sigmoid'))
  model.add(Dense(1))
    # Compile model
  model.compile(loss='mean_squared_error', optimizer='sgd')
  return model
```

```python
#Evaluate Model
estimators=[]
estimators.append(('standardize',StandardScaler()))
estimators.append(('mlp',KerasRegressor(build_fn=DL_model3,epochs=200, batch_size=20,verbose=0)))
pipeline=Pipeline(estimators)
kfold=KFold(n_splits=5,shuffle=True,random_state=0)
results=cross_val_score(pipeline,X,Y,cv=kfold,scoring='neg_root_mean_squared_error')
rmse=np.abs(results)
print("BiasError: %.2f (%.2f) VarError" % (rmse.mean(),rmse.std()))
```
